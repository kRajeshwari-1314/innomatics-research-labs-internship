# 1. User Login Check

## **Problem Statement:**

- Given a username and password, check whether login is successful.
- username = "admin"
- password = "1234"

## **Requirements:**

- Print "Login Successful" if both username and password match.
- Otherwise print "Invalid Credentials".

## **Real-World Application:**

- Authentication systems.


```python
# Predefined credentials
stored_username="admin"
stored_password="1234"

# Taking input from user
username=input("Enter your username:")
password=input("Enter your password:")

# Checking login condition
if username==stored_username and password==stored_password:
  print("Login Successfull")
else:
  print("Invalid credentials")

```

# 2. Pass / Fail Analyzer

## **Problem Statement:**

- Analyze the performance of students based on their marks.
- Given marks list: [45, 78, 90, 33, 60]

## **Requirements:**

- A student is considered **Pass** if marks â‰¥ 50.
- Count the total number of students who passed.
- Count the total number of students who failed.
- Print the final result clearly showing pass and fail counts.

## **Real-World Application:**

- Used in academic evaluation systems to determine student performance.


```python
# Pass / Fail Analyzer
Student_marks=[45,78,90,33,60]
pass_count=0
fail_count=0

for score in Student_marks:
  if score>=50:
    pass_count+=1
  else:
    fail_count+=1

print("Total student:",len(Student_marks))
print("Total number of passed students",pass_count)
print("Total number of failed students",fail_count)
```

# 3. Simple Data Cleaner

## **Problem Statement:**

- Clean and standardize a list of user names.
- Given list: [" Alice ", "bob", " CHARLIE "]

## **Requirements:**

- Remove extra spaces from the beginning and end of each name.
- Convert all names to lowercase.
- Print the cleaned and standardized list.

## **Real-World Application:**

- This process is used in data preprocessing before analysis to ensure consistency and accuracy in datasets.


```python
# Original list of names
names = [" Alice ", "bob", " CHARLIE "]

# clean the names
cleaned_names=[]
for item in names:
  new_item=item.strip() # remove extra spaces
  new_item=new_item.lower()#convert lower case
  cleaned_names.append(new_item)
print("cleaned names", cleaned_names)
```

# 4. Message Length Analyzer

## **Problem Statement:**

- Analyze the size of each message in a given list.
- Given messages: ["Hi", "Welcome to the platform", "OK"]

## **Requirements:**

- Print the length of each message.
- Flag messages that contain more than 10 characters.

## **Real-World Application:**

- Used in text filtering and validation systems to monitor message size limits.


```python
#List of messages
messages = ["Hi", "Welcome to the platform", "OK"]

#check each message
for message in messages:
  length=len(message)
  print("message:",message)
  print("length:",length)
  if length>10:
    print("Status:Flagged messages more than 10 characters")

```

# 5. Error Message Detector

## **Problem Statement:**

- Detect error messages from a list of system logs.
- Given logs: ["INFO", "ERROR", "WARNING", "ERROR"]

## **Requirements:**

- Count the number of "ERROR" entries.
- Print the total error count clearly.

## **Real-World Application:**

- Used in monitoring and log analysis systems to track system failures and issues.


```python
#error message detector
logs=["INFO", "ERROR", "WARNING", "ERROR"]
error_count=0
for log in logs:
  if log=="ERROR":
    error_count+=1
print("Total error count:",error_count)
```
